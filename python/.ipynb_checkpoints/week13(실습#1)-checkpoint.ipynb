```python
import csv
import matplotlib.pyplot as plt
f = open('seoul2.csv')
data = csv.reader(f)
next(data)
result=[]
result2=[]
for row in data:
    if row[-1] != '' :
        if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '01':
            result.append(float(row[-1]))
            result2.append(float(row[-2]))
plt.plot(result)
plt.plot(result2)
plt.show()
```
